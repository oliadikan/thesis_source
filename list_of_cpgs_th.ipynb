```python
fa = []

with open('/home/olyadikan/Documents/Thesis/GRCh38.p14.fasta', 'r') as fasta:
    line = fasta.readline()
    chrm = ''
    seq = ''
    while line != '':
        if line[0] == '>':
            if chrm != '':
                ind = []
                for l in range(len(seq)-1):
                    if seq[l] == 'C' and seq[l+1] == 'G':
                        ind.append(l)
                fa.append([chrm, ind])
                print(chrm)
                seq = ''
            chrm = line[1:]
        else:
            seq += line
        line = fasta.readline()
    ind = []
    for l in range(len(seq)-1):
        if seq[l] == 'C' and seq[l+1] == 'G':
            ind.append(l)
    fa.append([chrm, ind])
    print(chrm)
```

```python
fa = []

through = 0

st_e = []

with open('/home/olyadikan/Documents/Thesis/GRCh38.p14.fasta', 'r') as fasta:
    line = fasta.readline()
    chrm = ''
    seq = ''
    while line != '':
        if line[0] == '>':
            if chrm != '':
                ind = []
                start = through
                end = through + len(seq)-1
                st_e.append([chrm, [start, end]])
                for l in range(len(seq)-1):
                    if seq[l] == 'C' and seq[l+1] == 'G':
                        ind.append(l+through)
                fa.append([chrm, ind])
                print(chrm)
                through += len(seq)
                seq = ''
            chrm = line[1:]
        else:
            seq += line
        line = fasta.readline()
    ind = []
    start = through
    end = through + len(seq)-1
    st_e.append([chrm, [start, end]])
    for l in range(len(seq)-1):
        if seq[l] == 'C' and seq[l+1] == 'G':
            ind.append(l+through)
    fa.append([chrm, ind])
    print(chrm)
```
